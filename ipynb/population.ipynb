```python
import re

# 위도, 경도만 추출
with open(r"C:\Users\tmddn\Downloads\js_marker_data (2).txt", encoding='utf-8') as file:
    data = file.read()
    
matches = re.findall(r"lat:\s*([0-9.]+),\s*lng:\s*([0-9.]+)", data)

# 위도(lat), 경도(lng) 값만 추출{('')없이}
formatted_output = [f"{{ lat: {lat}, lng: {lng} }}" for lat, lng in matches]

output = ",\n".join(formatted_output)
print(output)
```

```python
import json

# 클러스터링 번호를 지정해 추출함
with open(r"C:\Users\tmddn\Desktop\donga5gi\population_project\public\clustered_output.json", encoding='utf-8') as file:
    data = json.load(file)

formatted_output = [
    {"lat": item["lat"], "lng": item["lng"], "cluster": item["cluster"]} for item in data
]
output_json = json.dumps(formatted_output, indent=6)
print(output_json)
```
