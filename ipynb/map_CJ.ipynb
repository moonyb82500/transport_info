```python
import pandas as pd
import numpy as np
```

```python
import folium
map = folium.Map(location=[36,127],zoom_start=7)
map
```

```python
CJ = pd.read_csv('CJ대한통운_위도경도.csv')
CJ
```

```python
for i in CJ.index:
    name = CJ.loc[i, '상호명']
    lat = CJ.loc[i, '위도']
    lon = CJ.loc[i, '경도']
    marker = folium.Marker([lat,lon], popup=name).add_to(map)
map
```

```python
map2 = folium.Map(location=[36, 127], zoom_start=7)
for i in CJ.index:
    name = CJ.loc[i, '상호명']
    lat = CJ.loc[i, '위도']
    lon = CJ.loc[i, '경도']
    if CJ['상호명'][i] == 'B':
        code_color = 'orange'
    else: code_color = 'black'
    marker = folium.Marker([lat,lon], popup=name, icon=folium.Icon(color=code_color)).add_to(map2)
map2
```
