```python
geolocator = Nominatim(user_agent="geoapiExercises")
```

```python
addresses = [
    '경기도 용인시 처인구 원삼면 원양로 487',
    '전북특별자치도 완주군 봉동읍 테크노밸리3로 65',
    '충청북도 음성군 삼성면 대성로 355',
    '광주광역시 광산구 평동산단외로 56',
    '대전광역시 대덕구 대전로1331번길 205',
    '경기도 용인시 처인구 백암면 죽양대로 1182',
    '충청남도 천안시 동남구 병천면 충절로 1529-5',
    '경기도 김포시 고촌읍 아라육로152번길 115',
    '경상남도 양산시 물금읍 제방로 245',
    '경상북도 경산시 와촌면 상암길37길 75-9',
    '충청북도 청주시 상당구 가덕면 상장인차로 188',
    '충청북도 청주시 서원구 남이면 저산척북로 203-44',
    '충청남도 당진시 송산면 가곡로 21'

]
# 위도와 경도 추출
locations = []
for address in addresses:
    try:
        location = geolocator.geocode(address)
        if location:
            locations.append((address, location.latitude, location.longitude))
        else:
            locations.append((address, None, None))
        time.sleep(5)
    except Exception as e:
        locations.append((address, None, None))
# 결과 출력
for loc in locations:
    print(f"Address: {loc[0]}, Latitude: {loc[1]}, Longitude: {loc[2]}")
```

```python
#5개 기업의 위,경도를 각자 전체 진행 후 none 값은 수기로 입력
```

```python
df_locations = pd.DataFrame(locations)
df_locations
```
